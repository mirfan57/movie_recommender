```python
# We want a list as ['Animation','Comedy','Family']. 
# But as our dictionary above is a string not a list we cannot loop through it.
# We have to convert this string of list of dictionary into a list of dictionary which can be done using ast.literal_eval function
```

```python
import ast
ast.literal_eval("[{'id': 16, 'name': 'Animation'}, {'id': 35, 'name': 'Comedy'}, {'id': 10751, 'name': 'Family'}]")
```

```python
import ast

def get_list(obj):
    my_list = []
    for i in ast.literal_eval(obj):       #literal_eval function converts the obj in string format to a list format
        my_list.append(i['name'])
    return my_list
```

```python
df['genres'] = df['genres'].apply(get_list)
df.head(2)
```

```python
df['keywords'][0]
```

```python
df['production_companies'] = df['production_companies'].apply(get_list)
df['keywords'] = df['keywords'].apply(get_list)
df.head(3)
```

```python
df['cast'][0]
```

We will only fetch the first 3 names in the cast column as they are probably important actors.

```python
def get_actors(obj):
    act = []
    count = 0
    for i in ast.literal_eval(obj):
        if count < 3:
            act.append(i['name'])
            count += 1
        else:
            break
    return act
```

```python
df['cast'] = df['cast'].apply(get_actors)
df.head(2)
```

```python
df['crew'][0]
```

```python
def get_director(obj):
    direct = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            direct.append(i['name'])
            break
    return direct
```

```python
df['crew'] = df['crew'].apply(get_director)
df.head(2)
```

Now we will concatenate all the lists to get a big paragraph which will act as our tag. We have to perform text vectorization, where similarity will be based on tags. We will convert the whole tags text in vector format and plot on a graph, all the different movies can be represented by vectors and the closeness is based on cosine distance.

We will concatenate all the tags to get a huge string and we will get top 10000 most common words. Then this will become our columns and all the individual movie tags will be our rows and the number of occurences of the given word in a particular tag will be entered in a row wise manner, where each row represents a vector of dimension **5,000*1** belonging to first movie. Likewise, we have in total 4763 movies so we get a table of **4763*5,000** dimensions. Plotting all these 4763 vectors in 5000 dimensional space and then fetching all the closest vectors to a given vector will give us the most closely associated movie to given movie.

We will not consider stop words during vectorization, which don't contribute in the sentence meaning. stop_words paramter will take care of this in the CountVectorizer class.

We see that many same words are repeated in different forms for example, adapt, adaptation and adapted are all same words. So, we need to get rid of this issue, we use ***stemming***.

```python
#!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Create a function stem which performs Stemming and gives us the root words
def stem_words(text):
    lst = []
    for i in text.split():
        lst.append(ps.stem(i))
    return " ".join(lst)
```

```python
final_df['tags'] = final_df['tags'].apply(stem_words)
final_df['tags']
```

### Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(final_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
vectors.shape
```

In such high dimensional data, Euclidean distance fails and it is not a good measure to calculate the closest distance, instead we use a more reliable measure called cosine distance which is prsent in scikit learn.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#type(vectors)
#vaex_vectors = vaex.from_arrays()
#vaex_vectors
```

```python
similarity = cosine_similarity(vectors)
```

```python
# it gives similarity of index 1 movie with all others. We see that at second position value is 1 because similarity with itself
similarity[1]
```

For getting top 10 closest movies to any given movie we have to sort the above obtained similarity array, but in this way we also loose the index positions of corresponding movies. So, we use enumerate to preserve this index position of sorted array.

```python
# we get the index positon and corresponding values
# now we want to sort but not on the basis of index positon but on the similarity metric, for this we use key=lambda x:x[1]
similarity_tuple = sorted(list(enumerate(similarity[1])), reverse=True, key=lambda x:x[1])
similarity_tuple
```

```python
def recommend_top_movies(movie):
    movie_index = final_df[final_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_tuple = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    
    for entry in movies_tuple:
        print(final_df.iloc[entry[0]].title)
```

```python
recommend_top_movies('Jumanji')
```
